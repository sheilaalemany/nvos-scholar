```python
def search_in_ontology(class_name, original_ontology):
    to_drop = []
    for initial in range(0, len(original_ontology)): 
        if class_name in original_ontology[initial]:
            final = initial+1
            if '<owl:equivalentClass>' in original_ontology[final]:   
                while '</owl:equivalentClass>' not in original_ontology[final]:
                    to_drop += [final] #index of places to drop starting at <owl:equivalentClass>
                    final+=1
                to_drop += [final] #drop to </owl:equivalentClass>
            if to_drop:
                del original_ontology[to_drop[0]: to_drop[len(to_drop)-1]+1]
                (initial, final) = search_in_ontology(class_name, original_ontology)
                    
            while '</owl:Class>' not in original_ontology[final]:
                final+=1

            return initial, final

def get_class_info(initial, final, original_ontology):
    class_to_replace = []
    for x in range(initial, final+1):
        class_to_replace += [original_ontology[x]]
    return class_to_replace

def save_class_info(f, initial, final, original_ontology):
    to_save = get_class_info(initial, final, original_ontology)
    f.write("\n".join(to_save))
    
```

```python
to_fill_lines = open('to-fill-in.rtf').read().splitlines()
```

```python
original_ontology = open('foodon-onotology-source.rtf').read().splitlines()
```

```python
f = open('final_save_again.txt', 'w')

for x in to_fill_lines:
    if '<owl:Class' in x:
        (initial, final) = search_in_ontology(x, original_ontology)
        save_class_info(f, initial, final, original_ontology)
        print('SAVED: '+x)

f.close()
```
